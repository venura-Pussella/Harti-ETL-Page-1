```python
import tabula
from tabulate import tabulate
import pandas as pd

# Function to read PDF and extract tables into a DataFrame
def extract_tables_from_pdf_to_df(pdf_path):
    tables = tabula.read_pdf(pdf_path, pages='2')
    
    # Combine all tables into a single DataFrame if there are multiple tables
    if isinstance(tables, list):
        combined_df = pd.concat(tables, ignore_index=True)
    else:
        combined_df = tables
    
    return combined_df

pdf_path = '../data/daily_16-07-2024.pdf'
1
food_df = extract_tables_from_pdf_to_df(pdf_path)

# Print the DataFrame in a tabular format for better readability
print(tabulate(food_df, headers='keys', tablefmt='psql'))

# Optionally save to CSV
# food_df.to_csv('food.csv', index=False)

food_df.head()
```

```python
# Identify columns with "."
columns_with_dot = [col for col in food_df.columns if '.' in col]

# Extract column names before "." and rename columns
new_column_names = {}
for col in columns_with_dot:
    new_col_name = col.split('.')[0]  # Get part before "."
    new_column_names[col] = new_col_name

# Rename columns
food_df.rename(columns=new_column_names, inplace=True)

# Print the updated DataFrame
food_df.head()
```

```python
# Get the name of the second column
second_column_name = food_df.columns[1]

# Rename the sixth column to match the name of the second column
food_df.rename(columns={food_df.columns[5]: second_column_name}, inplace=True)
food_df.rename(columns={food_df.columns[6]: second_column_name}, inplace=True)

food_df.head()
```

```python
cell_value1= food_df.iloc[0, 5]
cell2, cell3 = cell_value1.split()
print(cell2)
print(cell3)
```

```python
food_df.iloc[0, 6] = cell2
food_df.iloc[0, 5] = cell3
food_df.fillna(0, inplace=True)

# find this word in the variety column and replace it with "Karthakolomban"
food_df['Variety'] = food_df['Variety'].replace(['- Karathakolomb6a0n-', '- Karathakolomb5a0n-'], 'Karthakolamban')

food_df.head(60)
```

```python
# Transpose the DataFrame and set the first row as column headers
df_transposed = food_df.T
df_transposed.columns = df_transposed.iloc[0]

# Drop the first row (which is now redundant as column headers) and reset index
df_transposed = df_transposed.drop(df_transposed.index[0])
df_transposed = df_transposed.reset_index().rename(columns={'index': 'Date', 'Variety': 'Location'})

# Clean up the 'Date' column: remove trailing decimal and convert to datetime
df_transposed['Date'] = df_transposed['Date'].str.replace(r'\.\d+$', '', regex=True)
df_transposed['Date'] = pd.to_datetime(df_transposed['Date'])

# Reshape the DataFrame from wide to long format
df_melted = df_transposed.melt(id_vars=['Date', 'Location'], var_name='Item_Names', value_name='Value')
# Save the melted DataFrame to a CSV file
df_melted.to_csv('food1.csv', index=False)

df_melted.head(238)
```

```python
# Function to clean the value column
def clean_value(value):
    if isinstance(value, str) and value.startswith('-'):
        return value[1:]
    return value

# Search for 'Thambuththegama' in the 'location' column and clean the corresponding 'value' column
df_melted.loc[df_melted['Location'] == 'Thambuththegama', 'Value'] = df_melted.loc[df_melted['Location'] == 'Thambuththegama', 'Value'].apply(clean_value)

# Save the cleaned DataFrame to a new CSV file
df_melted.to_csv('cleaned_food1.csv', index=False)

df_melted.head()
```

```python
# Update Item_Names column
df_melted.loc[df_melted['Item_Names'].str.contains('- Small', na=False), 'Item_Names'] = 'Pineapple - Small'
df_melted.loc[df_melted['Item_Names'].str.contains('- Medium', na=False), 'Item_Names'] = 'Pineapple - Medium'

# Save the updated DataFrame to a new CSV file
df_melted.to_csv('updated_food1.csv', index=False)
```

```python
# Search for 'Peliyagoda' in the 'Location' column
location_mask = df_melted['Location'] == 'Peliyagoda'

# Within those rows, search for 'Lime' in the 'Item_Names' column
lime_mask = df_melted['Item_Names'].str.contains('Lime', na=False)

# Combine masks to find the relevant rows
combined_mask = location_mask & lime_mask

# Split the 'Item_Names' values by space for the relevant rows
lime_splits = df_melted.loc[combined_mask, 'Item_Names'].str.split(' ', expand=True)

# Assuming the last column of the split values is the third part
part1, part2, part3 = None, None, None

if len(lime_splits.columns) > 2:
    part1 = lime_splits.iloc[:, -3]
    part2 = lime_splits.iloc[:, -2]
    part3 = lime_splits.iloc[:, -1]

print(part1)
print(part2)
print(part3)

New_cell = part2 + ' ' + part3

print(New_cell)

```

```python
# Handling cells with just dashes or invalid patterns
df_melted['Value'] = df_melted['Value'].replace('-', '0 - 0')

# Split the 'Value' column into 'Min_Value' and 'Max_Value' columns
df_melted[['Min_Value', 'Max_Value']] = df_melted['Value'].str.split(' - ', expand=True)

# Convert the new columns to integers, handling potential conversion errors
df_melted['Min_Value'] = pd.to_numeric(df_melted['Min_Value'], errors='coerce').fillna(0).astype(int)
df_melted['Max_Value'] = pd.to_numeric(df_melted['Max_Value'], errors='coerce').fillna(0).astype(int)
```
